```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Google Web Graph

```python
data_networkx = pd.read_excel('./tables/networkx_benchmark/pagerank/networkx_pagerank_results_web.xlsx')
plt.plot(data_networkx ['Размер выборки'], data_networkx ['Время выполнения'],
         marker='s', label='Реализация netwokx')

data_cpp = pd.read_excel('./tables/my_algo/pagerank/my_cpp_pagerank_results_web.xlsx')
plt.plot(data_cpp['sample'], data_cpp['real_time'], label='Реализация на C++', marker='o')

data_python = pd.read_csv('./tables/my_algo/pagerank/my_py_pagerank_results_web.csv')
plt.plot(data_python['Размер выборки'], data_python['Время выполнения'],
         label='Реализация на Python', marker='*')

plt.xlabel('Доля набора данных')
plt.ylabel('Время выполнения, сек')


plt.grid(True)
plt.legend()
plt.show()
```
